```python
import pandas as pd
df_suicide = pd.read_csv('suicide_rate.csv')

# part 1

# avg number of suicides each year
avg_yearly_global_suicide = df_suicide['suicides_no'].mean()
print(avg_yearly_global_suicide)

# where and when max number of suicides in one year has occurred
max_number_suicide = df_suicide['suicides_no'].max()
max_suicide_row = df_suicide.loc[df_suicide['suicides_no']==max_number_suicide]


# total suicides by year
year_df = df_suicide.groupby(['year'])
tot_suicide_by_year = year_df['suicides_no'].sum()

# part 2

# number of suicides each year by sex
year_sex_df = df_suicide.groupby(['year', 'sex'])
suicide_by_year_sex = year_sex_df['suicides_no'].sum()

# number of suicides each year by generation
gen_df = df_suicide.groupby(['year','generation'])
suicide_by_year_gen = gen_df['suicides_no'].sum()

# part 3
# total average global suicides plotted over time
tot_suicide_by_year.plot(kind='line', title='Total Average Global Suicides Over Time', ylabel='Average Number Suicides')
```

```python
# global suicide rates by gender over time
suicide_by_year_sex.unstack().plot(title='Global Suicide Rates by Gender Over Time', ylabel='Number of Suicides')
```
